```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np

# Load Iris dataset
from sklearn.datasets import load_iris
iris_data = load_iris(as_frame=True)
iris_df = iris_data.frame

# Map target to species names
iris_df['species'] = iris_df['target'].map(dict(enumerate(iris_data['target_names'])))

# Set global Seaborn style
sns.set(style="whitegrid")

# Display the first few rows of the dataset
iris_df.head()
```

```python
# Plot distributions for each feature
features = iris_df.columns[:-2]  # Exclude target and species
plt.figure(figsize=(14, 10))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(iris_df[feature], kde=True, bins=20, color="c")
    plt.title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```

##This will show histograms and KDE (Kernel Density Estimate) curves for each feature in the dataset.

```python
# Pairplot to visualize pairwise relationships
sns.pairplot(iris_df, hue='species', palette='Set1', markers=["o", "s", "D"])
plt.show()
```

##This will create a pairplot showing pairwise relationships between all features, color-coded by species.

```python
# Boxplots for each feature across species
features = iris_df.columns[:-2]  # Exclude target and species
plt.figure(figsize=(14, 10))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='species', y=feature, data=iris_df, palette="Set2")
    plt.title(f'{feature} by Species')

plt.tight_layout()
plt.show()
```

##This will generate boxplots to observe the spread of each feature for each species

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
# Select only numeric features for correlation calculation
numeric_features = iris_df.select_dtypes(include=['number'])
corr_matrix = numeric_features.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Feature Correlation Heatmap')
plt.show()
```

##This will display a correlation matrix heatmap showing the correlation between each pair of features

```python
# Count plot for species distribution
plt.figure(figsize=(7, 5))
sns.countplot(x='species', data=iris_df, palette="coolwarm")
plt.title('Species Distribution')
plt.show()
```

```python
# Interactive scatter matrix using Plotly
fig = px.scatter_matrix(iris_df, dimensions=iris_df.columns[:-2], color="species",
                        title="Interactive Scatter Matrix",
                        labels={col: col.replace('_', ' ') for col in iris_df.columns[:-2]},
                        height=700)
fig.show()
```

##This creates an interactive scatter matrix, where you can hover over points and zoom in to explore relationships between features

```python
# Interactive 3D scatter plot using Plotly
fig = px.scatter_3d(iris_df, x='sepal length (cm)', y='sepal width (cm)', z='petal length (cm)',
                    color='species', symbol='species',
                    title="3D Scatter Plot of Sepal and Petal Dimensions")
fig.show()
```

##This creates an interactive 3D scatter plot where you can rotate and zoom to explore how features are distributed across the species

**Summary of Visualization Components:
Distribution Plots: Shows the distribution of individual features using histograms and KDE curves.
Pairwise Relationships: Pairplots display relationships between feature pairs, colored by species.
Boxplots: Shows feature spread and outliers per species.
Correlation Heatmap: A heatmap showing the correlation between different features.
Species Distribution: A bar plot to display the number of instances per species.
Interactive Plots (Plotly): Offers more interactivity with scatter matrices and 3D plots, useful for deep data exploration.**
